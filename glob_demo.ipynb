Tutorial 27 - Using glob to read multiple files in python
https://www.youtube.com/watch?v=Z90KEqJoC3w

```python
import glob
import cv2

file_list = glob.glob('images/test_images/*') #python list 
print(file_list)
print(type(file_list))

my_list = []
for file in file_list:
    print(file)
    a = cv2.imread(file)
    my_list.append(a)
    
from matplotlib import pyplot as plt
plt.imshow(my_list[2])
print(type(my_list[2]))
print(my_list[2].shape)
```
